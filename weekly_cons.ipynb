* Create a Time-Series Feature

```python
df['dow'] = df.index.dayofweek
df['doy'] = df.index.dayofyear
df['year'] = df.index.year
df['month'] = df.index.month
df['quarter'] = df.index.quarter
df['hour']=df.index.hour
df['weekday'] = df.index.weekday_name
df['woy']=df.index.weekofyear
df['dom']=df.index.day
df['date']=df.index.date
```

```python
_ = df[['PJM_Load', 'hour']].plot(x='hour',
                                  y='PJM_Load',
                                  kind='scatter',
                                  figsize=(14,4),
                                  title='Consumption by Hour of Day'
                                  )
```

```python
# figure out by the day of the week
_ = df.pivot_table(index=df['hour'],
                   columns="weekday",
                   values="PJME",
                   aggfunc='sum'
                   ).plot(figsize=(15,9),
                          title="PJM East - Daily Trends")
```
